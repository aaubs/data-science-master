# Activation Functions in PyTorch: A Comprehensive Guide

In this tutorial, we'll explore various activation functions available in PyTorch, understand their characteristics,
and visualize how they transform input data.

## Introduction

Activation functions are crucial in neural networks as they introduce non-linearity, allowing the network to solve complex problems
and make predictions or classifications.

## Types of Activation Functions



![](https://raw.githubusercontent.com/aaubs/ds-master/main/data/Images/activation_functions.png)

### 1. Sigmoid Function

The sigmoid function maps input values between 0 and 1, which is useful for binary classification or probability predictions.

```python
import torch
import torch.nn as nn

sigmoid = nn.Sigmoid()
input = torch.tensor([-2.0, -1.0, 0.0, 1.0, 2.0], dtype=torch.float32)
output = sigmoid(input)
print(output)
```

### Visualization of Sigmoid Function


```python
import matplotlib.pyplot as plt
import numpy as np

x = torch.linspace(-10, 10, 100)
y = sigmoid(x)

plt.plot(x.numpy(), y.numpy())
plt.title('Sigmoid Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

### 2. Tanh Function

The tanh function outputs values between -1 and 1.

```python
tanh = nn.Tanh()
output = tanh(input)
print(output)
```


### Visualization of Tanh Function


```python
y = tanh(x)

plt.plot(x.numpy(), y.numpy())
plt.title('Tanh Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

### 3. ReLU Function

The ReLU function outputs the input for all positive values, and zero for all negative values.

```python
relu = nn.ReLU()
output = relu(input)
print(output)
```

### Visualization of ReLU Function


```python
y = relu(x)

plt.plot(x.numpy(), y.numpy())
plt.title('ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

### 4. Leaky ReLU Function

Leaky ReLU allows a small, positive gradient when the input is negative.

```python
leaky_relu = nn.LeakyReLU()
output = leaky_relu(input)
print(output)
```

### Visualization of Leaky ReLU Function


```python
y = leaky_relu(x)

plt.plot(x.numpy(), y.numpy())
plt.title('Leaky ReLU Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

### 5. Softmax Function

The softmax function is used for multi-class classification problems.

```python
softmax = nn.Softmax(dim=0)
input = torch.tensor([2.0, 1.0, 0.1])
output = softmax(input)
print(output)
```

### Visualization of Softmax Function


```python
y = softmax(x)

plt.plot(x.numpy(), y.numpy())
plt.title('Softmax Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```

### 6. Identity Function

The Identity activation function is a linear activation function that outputs the input directly, without any transformation.

```python
identity = nn.Identity()
input = torch.tensor([2.0, 1.0, 0.1])
output = identity(input)
print(output)
```

### Visualization of Identity Function


```python
y = identity(x)

plt.plot(x.numpy(), y.numpy())
plt.title('Identity Activation Function')
plt.xlabel('Input')
plt.ylabel('Output')
plt.grid(True)
plt.show()
```


# Choosing the Right Activation Function

Choosing the right activation function for your neural network model is crucial.
Here are some general guidelines:

- Sigmoid and tanh functions are sometimes avoided due to the vanishing gradient problem.
- ReLU is a good default choice, but it should be noted that it’s not suitable for all kinds of data.
- If you have a problem with dead neurons in your network, Leaky ReLU or Parametric ReLU (PReLU) might help.
- Softmax is typically used in the output layer for classification problems.

The choice of activation function at the output layer is also crucial. For binary classification, a sigmoid function is typically used, while for multi-class classification, a softmax function is preferable.



# Conclusion

Activation functions are a fundamental component of neural networks. They introduce non-linearity to the system,
enabling the network to learn complex patterns and make predictions. Choosing the right activation function
for your model depends on the specific requirements of your task and the nature of your data.
