```python
# @title Plotting functions
def visualize_vectors(v, v_unit):
  """ Plots a 2D vector and the corresponding unit vector

  Args:
    v (ndarray): array of size (2,) with the vector coordinates
    v_unit (ndarray): array of size (2, ) with the unit vector coordinates

  """
  fig, ax = plt.subplots()

  # Set up plot aesthetics
  ax.spines['top'].set_color('none')
  ax.spines['bottom'].set_position('zero')
  ax.spines['left'].set_position('zero')
  ax.spines['right'].set_color('none')
  ax.set(xlim = [-6, 6], ylim = [-6, 6])
  ax.grid(True, alpha=.4, linewidth=1, zorder=0)

  # Plot vectors
  v_arr = ax.arrow(0, 0, v[0], v[1], width=0.08, color='#648FFF', length_includes_head = True, zorder = 2);
  v_unit_arr = ax.arrow(0, 0, v_unit[0], v_unit[1], width=0.08, color='#DC267F', length_includes_head = True, zorder = 3);
  ax.set(xlim = [-4, 4], ylim = [-4, 4]);

  # Add legend
  leg = ax.legend([v_arr, v_unit_arr], [r"Vector $\mathbf{v}$", r"Normalized vector $\tilde{\mathbf{v}}$"], handlelength = 0, fontsize = 20, loc = 'upper left')
  for handle, label in zip(leg.legendHandles, leg.texts):
    label.set_color(handle.get_facecolor())
    handle.set_visible(False)

from matplotlib.patches import FancyArrowPatch
from mpl_toolkits.mplot3d import proj3d

class Arrow3D(FancyArrowPatch):
    def __init__(self, xs, ys, zs, *args, **kwargs):
        FancyArrowPatch.__init__(self, (0,0), (0,0), *args, **kwargs)
        self._verts3d = xs, ys, zs

    def draw(self, renderer):
        xs3d, ys3d, zs3d = self._verts3d
        xs, ys, zs = proj3d.proj_transform(xs3d, ys3d, zs3d, renderer.M)
        self.set_positions((xs[0],ys[0]),(xs[1],ys[1]))
        FancyArrowPatch.draw(self, renderer)
```

# How to Calculate Length Using L1 and L2 Norms in Python
In this tutorial, we will calculate the length (or norm) of a vector with two elements using both L1 norm and L2 norm in Python. This will help you understand how to use NumPy to compute these norms efficiently.

### 1. L1 Norm (Manhattan Norm)

#### Formula:
For a vector $x = [x_1, x_2, \dots, x_n]$, the **L1 norm** is calculated as:

$
||x\||_1 = |x_1| + |x_2| + \dots + |x_n|
\$

This represents the sum of the absolute values of the vector components.

#### Example:
Given $x = [3, 4] $, compute the L1 norm.





$||x\||_1 = |3| + |4| = 3 + 4 = 7
\$

So, the L1 norm of vector $x$ is **7**.

### 2. L2 Norm (Euclidean Norm)

#### Formula:
For a vector $x = [x_1, x_2, \dots, x_n]$, the **L2 norm** is calculated as:

$
\||x\||_2 = \sqrt{x_1^2 + x_2^2 + \dots + x_n^2}
\$

This represents the square root of the sum of the squares of the vector components.

#### Example:
Given $ x = [3, 4] $, compute the L2 norm.

$
\||x\||_2 = \sqrt{3^2 + 4^2} = \sqrt{9 + 16} = \sqrt{25} = 5
\$

So, the L2 norm of vector \( x \) is **5**.

## The Jaccard Similarity

The Jaccard similarity is a measure used to compare the similarity between two sets. It is defined as the size of the intersection divided by the size of the union of the sets. You can represent the Jaccard similarity between sets \( A \) and \( B \) as:

$$ J(A, B) = \frac{|A ∩ B|}{|A ∪ B|} $$

### Steps to Compute Jaccard Similarity

1. **Identify the Sets**: Start by identifying the two sets \( A \) and \( B \) that you want to compare.
2. **Calculate the Intersection**: Compute the intersection \( A ∩ B \), which consists of the elements that are common to both \( A \) and \( B \).
3. **Calculate the Union**: Compute the union \( A ∪ B \), which consists of all the unique elements from both \( A \) and \( B \).
4. **Compute the Jaccard Similarity**: Use the formula to calculate \( J(A, B) \).


### Example of Jaccard Similarity with Nominal Sets

In the case of nominal sets, the elements are categorical rather than numerical. For example, let's say we have two sets of fruits:

$$ A = \{ "Apple", "Banana", "Cherry" \} $$
$$ B = \{ "Banana", "Cherry", "Ananas" \} $$

### Calculating Jaccard Similarity for Nominal Sets

- **The Intersection**:
$$  A ∩ B = \{ "Banana", "Cherry" \} $$
- **The Union**:
$$  A ∪ B = \{ "Apple", "Banana", "Cherry", "Ananas" \} $$  
- **The Jaccard Similarity**:
$$  J(A, B) = \frac{2}{4} = 0.5 $$
