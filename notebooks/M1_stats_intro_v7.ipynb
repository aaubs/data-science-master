
# Probability Theory 101

Probability is a measure that quantifies the likelihood of an event occurring. In the context of statistics and data science, understanding probability is essential for making predictions and informed decisions.

Learning probability is crucial for both descriptive and inferential statistics because it provides the foundational framework and methodologies that enable you to understand, describe, and make predictions about data.

## Basic Concepts of Probability

- **Experiment:** An experiment in probability is a process or action you carry out that can have different results each time. For example, rolling a die is an experiment because you might roll any number from 1 to 6, and you're not sure what you'll get each time you roll it.

- **Sample Space (S):** The complete set of potential outcomes from rolling the die. In this case, the sample space is \( S = \{1, 2, 3, 4, 5, 6\} \).

- **Event:** Refers to any specific outcome or combination of outcomes from the sample space. An example of an event is rolling a "5."

- **Probability of an Event (P(E)):** Represented by \( P(E) \), where \( E \) denotes an event. This measure reflects the chance of an event occurring, with possible values ranging from 0 to 1. For a fair die, the probability that a "5" appears is \( P(E) =  1/6\).

These foundational concepts are crucial for analyzing the results of multiple dice rolls, allowing for predictions regarding the frequency of different outcomes.


These principles help in understanding how often different outcomes are expected to occur when the coin is tossed repeatedly.


# Sampling and Central Limit Theorem

## Sampling

Sampling is the process of selecting a subset (a statistical sample) of individuals from a statistical population to estimate characteristics of the whole population. Proper sampling methods ensure that the sample is representative of the entire population.

Types of Sampling:
- **Random Sampling**: Every individual has an equal chance of being selected.
- **Stratified Sampling**: The population is divided into subgroups (strata) and random samples are taken from each stratum.
- **Cluster Sampling**: The population is divided into clusters, and a random sample of clusters is selected.

## Central Limit Theorem (CLT)

The Central Limit Theorem is a fundamental theorem in probability theory and statistics. It states:

***The distribution of sample means will look like a normal distribution, no matter what the original population's distribution is, provided that the sample size is sufficiently large.***

Properties:
- The mean of the sample mean is equal to the population mean.
- The standard deviation of the sample mean is equal to the population standard deviation divided by the square root of the sample size.


```python
# The population mean
data['age'].mean()
```

## 4. Measures of Position


The **Interquartile Range**, commonly denoted as IQR, is a measure of statistical dispersion or spread, representing the difference between the upper and lower quartiles. In simpler terms, IQR indicates the range within which the central 50% of values lie.

1. **Robustness**: Unlike range (which considers the absolute highest and lowest values), IQR only looks at the spread of the middle 50%, making it less susceptible to outliers and extreme values.
2. **Spread of Central Data**: While measures like variance or standard deviation provide an overall spread, IQR specifically gives insight into the spread of the central 50% of data.
3. **Identifying Outliers**: The IQR can be used to identify potential outliers. Any value that is below $( Q1 - 1.5 \times \text{IQR} )$ or above $( Q3 + 1.5 \times \text{IQR} )$ is generally considered an outlier.

Where:
- $( Q1 )$ represents the first quartile or the 25th percentile.
- $( Q3 )$ represents the third quartile or the 75th percentile.

In essence, the IQR provides a measure of where the bulk of the values lie in a dataset, ignoring extreme values.

```python
# Calculate Q1 and Q3
Q1_age = data['age'].quantile(0.25)
Q3_age = data['age'].quantile(0.75)
```

### A. Identifying Outliers Using Z-score

The Z-score is a numerical measurement that describes a value's relationship to the mean of a group of values. It's measured in terms of standard deviations from the mean. If a Z-score is 0, it indicates that the data point's score is identical to the mean score. A Z-score of 1.0 indicates a value that is one standard deviation from the mean, and so on.


$[Z = \frac{X - \mu}{\sigma}]$

Where:
- $Z$ is the Z-score.
- $X$ is the value of the data point.
- $\mu$ is the mean of the dataset.
- $\sigma$ is the standard deviation of the dataset.

Typically, data points with a Z-score value greater than 3 or less than -3 are considered outliers. This is based on the empirical rule (or 68-95-99.7 rule) which states that:
- About 68% of the data falls within one standard deviation.
- About 95% falls within two standard deviations.
- About 99.7% falls within three standard deviations.

So, data points that fall beyond three standard deviations can be considered rare or outliers.

### 2. Encoding
Encoding is the process of converting categorical variables into numerical format, which is required for most machine learning algorithms.


- Label Encoding: Converts each category into a unique number. This method is useful for ordinal categorical variables where the order matters, but it can introduce unintended ordinal relationships for non-ordinal data.

- One-Hot Encoding: Creates binary columns for each category, where 1 indicates the presence of a category and 0 indicates its absence. This method is useful for non-ordinal categorical variables with no intrinsic order but can lead to high-dimensional data if there are many categories.

- Dummy Encoding: A variant of One-Hot Encoding, Dummy Encoding creates binary columns for each category but drops one category to avoid multicollinearity. This method is particularly useful in regression models where including all categories might lead to redundancy and incorrect results.

| **Encoding Type**     | **Description**                                                                 | **Example for `Pclass` or `Sex` Feature**   | **Use Case**                                                      |
|-----------------------|---------------------------------------------------------------------------------|---------------------------------------------|-------------------------------------------------------------------|
| **Label Encoding**    | Assigns a unique integer to each category.                                       | `Pclass`: 1 -> 1, 2 -> 2, 3 -> 3            | Ordinal categorical variables (e.g., education)                    |
| **One-Hot Encoding**  | Creates a binary column for each category, indicating the presence of the category with 1. | Two columns: `male` -> [1, 0], `female` -> [0, 1] | Non-ordinal categorical variables with many unique categories     |
| **Dummy Encoding**    | Similar to One-Hot Encoding but drops one category to avoid multicollinearity.  | `Sex_male` column: 1 = male, 0 = female     | Categorical variables used in regression models to prevent multicollinearity |

## **Bivariate Analysis**

### Pearson's Correlation Coefficient
- **Definition**: Pearson's correlation coefficient measures the strength and direction of the linear relationship between two continuous variables.
- **Symbol**: Denoted by \( r \).
- **Range**: Ranges from -1 to 1, where:
  - \( r=1 \) indicates a perfect positive linear relationship.
  - \( r=-1 \) indicates a perfect negative linear relationship.
  - \( r=0 \) suggests no linear relationship.
- **Assumptions**:
  - Assumes both variables are normally distributed.
  - Assumes the relationship is linear.
- **Sensitivity**: Sensitive to outliers, which can significantly influence the correlation value.

### Spearman's Rank Correlation Coefficient
- **Definition**: Measures the strength and direction of a monotonic relationship between two variables, using ranks.
- **Non-parametric**: Does not assume a normal distribution of the data.
- **Applicability**: Useful for ordinal data or non-normally distributed data.
- **Range**: Also ranges from -1 to 1, where:
  - \( r=1 \) implies a perfect positive monotonic relationship.
  - \( r=-1 \) implies a perfect negative monotonic relationship.
  - \( r=0 \) indicates no monotonic relationship.



![](https://statistics.laerd.com/statistical-guides/img/spearman-1-small.png)

```python
# Calculate Pearson's correlation coefficient between 'Age' and 'Fare'
pearson_corr = data['age'].corr(data['fare'], method='pearson')
print(f"Pearson's Correlation Coefficient between Age and Fare: {pearson_corr:.2f}")

# Calculate Spearman's rank correlation coefficient between 'Age' and 'Fare'
spearman_corr = data['age'].corr(data['fare'], method='spearman')
print(f"Spearman's Rank Correlation Coefficient between Age and Fare: {spearman_corr:.2f}")
```

## **Univariate analysis**



## Confidence Intervals (CI)

### Overview
A confidence interval provides a range that likely includes a population parameter, such as the mean, based on data from a sample. It helps express the uncertainty of an estimate from sample data.

### Key Concepts
- **Point Estimate**: This is typically the sample mean, used as the best single guess for the population mean.
- **Margin of Error**: This indicates the extent around the point estimate where the true population parameter is expected to lie.

### Margin of Error Calculation
For a 95% confidence interval of a sample mean from a normally-distributed population, the formula is:

$\bar{x} \pm 1.96 \times \left(\frac{\sigma}{\sqrt{n}}\right)$

Where:
- $\bar{x}$ is the sample mean.
- $\sigma$ is the population standard deviation.
- $n$ is the sample size.
- The number 1.96 represents the Z-score that corresponds to a 95% confidence level.

#### Standard Error of the Mean (SEM)

The **Standard Error of the Mean (SEM)** is a statistical measure that quantifies the amount of variability or dispersion of sample means around the true population mean. In simpler terms, it tells us how much the sample mean (average) is expected to vary from the actual population mean.
  \
  $
   \text{SEM} = \frac{\sigma}{\sqrt{n}}
  $
   - **$\sigma\$**: The standard deviation of the population.
   - *n*: The sample size.


#### Interpretation
A 95% confidence interval means that if we repeat the sampling process many times, approximately 95% of the intervals we create will contain the actual population mean. Essentially, it gives us a way to predict the range within which the true mean likely falls, with a known level of confidence.

If we calculate a 95% confidence interval for the average height of a group based on a sample, this interval will likely include the true average height of the entire group in 95 out of 100 cases of repeated sampling.


### Practical Example

# Hypothesis Testing

Hypothesis testing is a statistical method used to make decisions about populations based on sample data. It's a fundamental concept in inferential statistics.

## Steps in Hypothesis Testing:

1. **Setting Hypotheses**:
    - ( $ H_0 $): Null hypothesis (a statement that indicates there is no effect or no difference in the data).
    - ($ H_1 $): Alternative hypothesis (what you want to prove).

2. **Choose a Significance Level (($ \alpha $))**: Typically set at 0.05. *It shows the probability of incorrectly rejecting the null hypothesis when it's actually true.*

3. **Collect and Analyze Data**: Calculate a test statistic based on sample data.

4. **Decision Making**: Compare the test statistic to a critical value and decide whether to reject the null hypothesis.

## 1. Hypothesis Testing (One sample T-Test, Two samples T-Test, and ANOVA)

**Objective:** To compare the mean fare paid by passengers who survived the Titanic disaster with those who did not.


**Hypothesis Statement:**
- **Null Hypothesis (\(H_0\))**: The mean fare paid by survivors is equal to the mean fare paid by non-survivors.
  $
  H_0: \mu_{\text{survivor}} - \mu_{\text{nonsurvivor}} = 0
  $
  Where:
  - $\mu_{\text{survivor}}$ is the mean fare paid by survivors.
  - $\mu_{\text{nonsurvivor}}$ is the mean fare paid by non-survivors.

- **Alternative Hypothesis (\(H_A\))**: The mean fare paid by survivors is different from the mean fare paid by non-survivors.
$
  H_A: \mu_{\text{survivor}} - \mu_{\text{nonsurvivor}} \neq 0
 $

This is a two-tailed test because we are interested in detecting any significant difference in either direction (i.e., whether survivors pay more or less on average compared to non-survivors).



![](https://miro.medium.com/v2/resize:fit:1400/format:webp/1*VXxdieFiYCgR6v7nUaq01g.jpeg)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats


# Hypothesis: Compare the mean fare paid by passengers who survived and those who did not survive.
# Extract fare data for the two groups: survivors and non-survivors
survivor_fares = data[data['survived'] == 1]['fare'].dropna()
nonsurvivor_fares = data[data['survived'] == 0]['fare'].dropna()

# Sample 30 fares from each group
np.random.seed(42)
sample_size = 30
survivor_sample = survivor_fares.sample(sample_size, random_state=42)
nonsurvivor_sample = nonsurvivor_fares.sample(sample_size, random_state=42)

# Perform a t-test
t_stat, p_value = stats.ttest_ind(survivor_sample, nonsurvivor_sample)
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
# Calculate the observed mean difference
mean_diff = survivor_sample.mean() - nonsurvivor_sample.mean()

# Calculate standard error of the difference in means
std_survivor = survivor_sample.std(ddof=1)
std_nonsurvivor = nonsurvivor_sample.std(ddof=1)
se_diff = np.sqrt((std_survivor**2 / sample_size) + (std_nonsurvivor**2 / sample_size))

# Plot the null hypothesis distribution and rejection regions
plt.figure(figsize=(20, 6))

# Null hypothesis mean difference (H0: mean_survivor - mean_nonsurvivor = 0)
null_mean_diff = 0

# Adjusting the x range if needed based on the distribution
x_range = max(abs(mean_diff) + 10, 30)  # Adjust 10 and 30 based on your data context
x = np.linspace(-x_range, x_range, 1000)
y = stats.norm.pdf(x, loc=null_mean_diff, scale=se_diff)
plt.plot(x, y, label='Null Hypothesis Distribution (H0: mean_survivor - mean_nonsurvivor = 0)', color='blue')

# Mark the observed mean difference
plt.axvline(mean_diff, color='red', linestyle='--', label=f'Observed Mean Difference = {mean_diff:.2f}')

# Shade the rejection region
alpha = 0.05
crit_value = stats.norm.ppf(1 - alpha / 2) * se_diff  # Scale critical value by the std deviation of the difference
plt.fill_between(x, 0, y, where=(x > crit_value) | (x < -crit_value), color='grey', alpha=0.1, label=f'Rejection Region (α = {alpha})')

# Add labels and legend
plt.title('Null Hypothesis Rejection Region (Titanic Fare Data)')
plt.xlabel('Difference in Means')
plt.ylabel('Density')
plt.legend()

# Show the plot
plt.show()

# Output the p-value for context
print("P-value:", p_value)
```

The evidence supports the alternative hypothesis that there is a difference in the mean fares paid by passengers who survived versus those who did not. A P-value of approximately 0.014 means there's a 1.414% chance of observing the data—or something more extreme—if the null hypothesis were true. The probability that this noticeable difference (or a more extreme difference) would occur just by random chance, if there really were no difference at all, is very small — only 1.4%. This means that the probability that the observed result of the test could occur by random chance, if the null hypothesis were true, is 1 percent.

```python
# Creating a contingency table of survival and sex (to display the frequency distribution of variables)
contingency_table = pd.crosstab(data['survived'], data['sex'])

# Performing the Chi-squared test
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-squared Statistic:", chi2)
print("P-value:", p)
```

## Understanding Odds and Logistic Regression

### Odds
- **Definition**: The odds of an event occurring is the ratio of the probability of the event happening to the probability of it not happening.
  
  $
  \text{Odds of Survival} = \frac{P(\text{Survived} = 1)}{P(\text{Survived} = 0)}
  $

### Log-Odds
- **Definition**: Logistic regression models the log-odds (logarithm of odds) of the dependent variable as a linear combination of predictor variables.
  
  $
  \log\left(\frac{P(\text{Survived} = 1)}{P(\text{Survived} = 0)}\right) = \beta_0 + \beta_1 \cdot \text{Age} + \beta_2 \cdot \text{Pclass}
  $
  
  - $\beta_0$ is the intercept.
  - $\beta_1$ and $\beta_2$ are the coefficients for Age and Passenger Class, respectively.

### Odds Ratio
- **Definition**: The odds ratio is obtained by exponentiating the coefficients of the logistic regression model. It shows how the odds of the dependent variable change with a one-unit increase in the predictor variable.
  
  $
  \text{Odds Ratio} = e^{\beta}
 $
  
  - For example, if the coefficient for Age is $\beta_1$, then $e^{\beta_1}$ represents the change in odds of survival for each additional year of age.

```python
# 1. Linearity
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.stats.diagnostic import het_breuschpagan
from scipy.stats import shapiro

# Assuming 'results' is the fitted OLS model from statsmodels
residuals = results.resid
fitted_vals = results.fittedvalues

# Linearity and Homoscedasticity
plt.scatter(fitted_vals, residuals)
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('Fitted Values')
plt.ylabel('Residuals')
plt.title('Residual vs. Fitted Plot')
plt.show()
```

In the plot, most residuals center around zero for lower fitted values, suggesting the model fits well there.

If there is a specific pattern or shape (like a funnel or curve), this suggests that the model might be missing some explanatory information, or there might be issues like heteroscedasticity (non-constant variance of residuals).

```python
# 2. Normality of Residuals
fig, ax = plt.subplots(1,2, figsize=(12, 4))
sns.histplot(residuals, kde=True, ax=ax[0])
sm.qqplot(residuals, line='s', ax=ax[1])
plt.show()

# Normality test of residuals
stat, p = shapiro(residuals)
print('Shapiro-Wilk Test p-value: {:.10f}'.format(p))
```

Since the p-value is much smaller than 0.05, we have strong evidence to reject the null hypothesis (The null hypothesis for the Shapiro-Wilk test is that the data is normally distributed). This means that the data does not follow a normal distribution.





```python
# 3. Homoscedasticity
_, pval, _, _ = het_breuschpagan(residuals, results.model.exog)
print('Breusch-Pagan p-value: {:.10f}'.format(pval))
```

Since the p-value is much smaller than 0.05, there is strong evidence to reject the null hypothesis, which states that the variance of the errors is constant (homoscedasticity). This means that the data exhibits heteroscedasticity, indicating that the variance of the errors is not constant.

```python
# 4. Independence of Residuals
print("Durbin-Watson statistic:", sm.stats.durbin_watson(residuals))
```

This value is very close to 2, which suggests that there is little to no autocorrelation in the residuals of the regression model. Essentially, the residuals appear to be independent, which is a good sign for the validity of the regression model.

```python
# 5. No Multicollinearity
from statsmodels.stats.outliers_influence import variance_inflation_factor
predictors = results.model.exog
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(predictors, i) for i in range(predictors.shape[1])]
vif["features"] = results.model.exog_names
print(vif)
```

## **Exercise 12:** Analysis of Factors Influencing Airbnb Listing Prices
#### Objective:
To understand how the number of bedrooms, review scores, and the number of reviews influence the log-transformed price of Airbnb listings.

#### Hypothesis:
- Null Hypothesis (H0): The number of bedrooms, review scores value, and the number of reviews have no significant effect on the log-transformed price of Airbnb listings.
- Alternative Hypothesis (H1): At least one among the number of bedrooms, review scores value, and the number of reviews has a significant effect on the log-transformed price of Airbnb listings

In the plot, most residuals center around zero for lower fitted values, suggesting the model fits well there.

**Number of Reviews:** The coefficient for the number of reviews is -0.001, which is statistically significant at the 1% level (indicated by three asterisks). This suggests that with every additional review, the logarithm of the price decreases by 0.001, all else being equal. This could imply that listings with more reviews might be slightly cheaper, or perhaps that more frequently reviewed properties tend to be more competitively priced.

**Review Scores Value:** The coefficients for review scores value are -0.058 and -0.059 in Models 2 and 3, respectively, each significant at the 1% level. This indicates that a one-point increase in review score value is associated with a decrease in the logarithm of the price by approximately 0.058 or 0.059. This relationship remains consistent across both models in which the variable appears, suggesting a robust inverse relationship between review scores and price. This might suggest that higher-rated properties (in terms of guest satisfaction) are priced lower, potentially as a strategy to attract more bookings.

To calculate the change in price for each unit change in the number of reviews and review scores, we need to convert the regression coefficients from the log price to actual price. The regression coefficients affect the logged price, but we can translate these effects to the original price scale using the following formula:

$Change_{inprice} = (e^{\text{coefficient}} - 1) \times 100\% $

Where $ e $ is the base of the natural logarithm. This formula calculates the percentage change in price for a one-unit increase in the respective independent variable.


```python
import math

# Coefficients for number of reviews and review scores
coeff_number_of_reviews = -0.001
coeff_review_scores = -0.059

# Calculating the percentage change in price for each coefficient
change_price_reviews = (math.exp(coeff_number_of_reviews) - 1) * 100
change_price_scores = (math.exp(coeff_review_scores) - 1) * 100

change_price_reviews, change_price_scores
```

- For each additional review: The price is expected to decrease by approximately 0.10%.
- For each point increase in review scores: The price is expected to decrease by approximately 5.73%.

**In general, based on our data, we can say that the pricing strategy plays an important role in attracting customers.**
